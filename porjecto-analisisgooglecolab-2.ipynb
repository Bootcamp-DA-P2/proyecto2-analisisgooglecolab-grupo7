# **Limpieza de datos**

```python
#Fechas
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['funded_time'] = pd.to_datetime(df['funded_time'], errors='coerce')
#Eliminar duplicados
df = df.drop_duplicates()
#Valores en blanco
# Eliminar filas sin información
df = df.dropna(subset=['loan_amount', 'country'])
# Rellenar valores en blanco
df['sector'] = df['sector'].fillna('Sin datos')
```

# **Limpieza de datos**

```python
#Fechas
fechas = [
    "posted_time",
    "disbursed_time",
    "funded_time",
    "date"
]

for col in fechas:
    df[col] = pd.to_datetime(df[col], errors="coerce")

df[fechas].info()
#Eliminar duplicados
df = df.drop_duplicates()
#Valores en blanco
# Eliminar filas sin información
df = df.dropna(subset=['loan_amount', 'country'])
# Rellenar valores en blanco
df['sector'] = df['sector'].fillna('Sin datos')
```

# **Limpieza de datos**

```python
#Fechas
fechas = ["posted_time","disbursed_time","funded_time","date"]

for col in fechas:
    df[col] = pd.to_datetime(df[col], errors="coerce")

df[fechas].info()
#Eliminar duplicados
df = df.drop_duplicates()
#Valores en blanco
# Eliminar filas sin información
df = df.dropna(subset=['loan_amount', 'country'])
# Rellenar valores en blanco
df['sector'] = df['sector'].fillna('Sin datos')
#Corregir la columna borrower_genders
df["borrower_genders"] = df["borrower_genders"].str.split(",").str[0].str.strip()
#Eliminamos la columna tags
df = df.drop(columns=["tags"])
```

# **Normalizar textos**

```python
df.head(5)
```

```python

#Valores atípicos
import matplotlib.pyplot as plt
numericas = ["funded_amount", "loan_amount"]

df[numericas].boxplot(figsize=(12,6))
plt.show()
```

```python
# borrower_genders reemplazar por "Sin_informacion"
df["borrower_genders"] = df["borrower_genders"].fillna("Sin_informacion")

#Errores por alguna mayuscula
texto_cols = ["activity","sector","country","region","currency","repayment_interval"]
for col in texto_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

# **Correccion de formato numericos**

```python
# Columnas de DINERO (€)
columnas_dinero = ["funded_amount","loan_amount"]
for col in columnas_dinero:
    df[col] = pd.to_numeric(df[col], errors="coerce")


# Numeros enteros

columnas_enteras = ["id","partner_id","term_in_months","lender_count"]
for col in columnas_enteras:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col] = df[col].astype("Int64")

df.head(5)
```

# **Creacion de columnas utiles**

```python
# Crear columnas separadas
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day

col = "loan_amount"

def clasificar_prestamo(valor):
    if pd.isna(valor):
        return "desconocido"
    elif valor < 500:
        return "pequeño"
    elif valor <= 2000:
        return "mediano"
    else:
        return "grande"

df["loan_size"] = df[col].apply(clasificar_prestamo)

df.head(5)
```

```python

# 1. Contar préstamos por sector
conteo_sectores = df['sector'].value_counts().nlargest(10)

# 2. Crear el gráfico de pastel
plt.figure(figsize=(10,7))
# 'autopct' muestra el porcentaje, 'startangle' gira el gráfico para mejor vista
conteo_sectores.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='Paired')

plt.title('Distribución de Préstamos por Sector')
plt.ylabel('')  # Elimina la etiqueta lateral 'sector'
plt.axis('equal')
plt.show()

# 3. Mostrar la tabla resumen
print(conteo_sectores)

```

```python
# 1. Función para limpiar y categorizar el género
def categorizar_genero(genero_str):
    if pd.isna(genero_str): 
        return "No especificado"
    
    # Convertimos a minúsculas y separamos por coma
    lista_generos = [g.strip().lower() for g in str(genero_str).split(',')]
    generos_unicos = set(lista_generos)
    
    if len(generos_unicos) == 1:
        # Si todos en el grupo son del mismo género
        return "Femenino" if "female" in generos_unicos else "Masculino"
    else:
        # Si hay mezcla de hombres y mujeres en el grupo
        return "Mixto"

# Aplicamos la función para crear una nueva columna limpia
df['genero_limpio'] = df['borrower_genders'].apply(categorizar_genero)

# 2. Contar préstamos por la nueva categoría de género
conteo_generos = df['genero_limpio'].value_counts()

# 3. Crear el gráfico de pastel
plt.figure(figsize=(10, 7))
# Usamos colores representativos: Rosado para mujeres, Azul para hombres, Verde para mixto
colores = ['#ff9999','#66b3ff','#99ff99', '#ffcc99']

conteo_generos.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colores,
    explode=[0.05] * len(conteo_generos) # Separa un poco las tajadas
)

plt.title('Distribución de Préstamos por Género del Prestatario')
plt.ylabel('') 
plt.axis('equal')
plt.show()

# 5. Mostrar la tabla resumen
print("Resumen de préstamos por género:")
print(conteo_generos)
```

```python
# 1. Obtener el Top 10 de países con más préstamos
top_paises = df['country'].value_counts().nlargest(10)

# 2. Crear el gráfico de barras horizontales
plt.figure(figsize=(12, 8))

# Usamos colores degradados para que se vea más profesional
top_paises.sort_values().plot(kind='barh', color='skyblue', edgecolor='navy')

# 3. Personalización del gráfico
plt.title('Top 10 Países con Mayor Cantidad de Préstamos', fontsize=16, fontweight='bold')
plt.xlabel('Número de Préstamos', fontsize=12)
plt.ylabel('País', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7) # Añade rejilla solo en el eje X

# Añadir los valores al final de cada barra para mayor claridad
for index, value in enumerate(top_paises.sort_values()):
    plt.text(value, index, f' {value}', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

# 4. Mostrar tabla resumen
print("Ranking de Préstamos por País:")
print(top_paises)
```
